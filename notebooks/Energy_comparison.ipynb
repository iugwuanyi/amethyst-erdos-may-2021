```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os.path

DATA_DIR = os.path.join('..','data')
```

```python
extracted_data = pd.read_csv(os.path.join(DATA_DIR,'All_Extracted_Data_From_PDFs.csv'))
```

```python
extracted_data = extracted_data[extracted_data['Topic'] == 'Energy and Grid']
extracted_data.head()
```

```python
energy_grid_by_country = extracted_data[['Country_Code','Amount_in_usd']].groupby('Country_Code').sum('Amount').reset_index()
energy_grid_by_country
```

```python
per_capita_access = pd.read_csv(os.path.join(DATA_DIR,'world_bank_access_to_electricity_per_pop.csv'))
per_capita_access
```

This data is wide with columns for each year. To join with our energy grid loans by year and country, we need to un-pivot this table and rename the country code:

```python
per_capita_access_unpivoted = per_capita_access.melt(
    id_vars = ['Country Name','Country Code','Indicator Name'],
    var_name = 'Year',
    value_name = 'Percent Access'
)
per_capita_access_unpivoted.rename(columns={'Country Code':'Country_Code'},inplace=True)
per_capita_access_unpivoted = per_capita_access_unpivoted[['Country_Code','Year','Percent Access']]
per_capita_access_unpivoted.sample(10,random_state=8001)
```

We'll linearly interpolate missing values.

```python
per_capita_access_unpivoted = per_capita_access_unpivoted.sort_values('Year').groupby('Country_Code').apply(lambda group: group.interpolate(limit_direction='both')).reset_index()
per_capita_access_unpivoted
```

To create a delta model from 1990 to 2020, we need to re-pivot:

```python
per_capita_access_interpolated = per_capita_access_unpivoted.pivot_table(index='Country_Code',columns='Year',values='Percent Access')
per_capita_access_interpolated = per_capita_access_interpolated.dropna().reset_index()
per_capita_access_interpolated
```

To calculate the access deltas:

```python
access_delta = pd.DataFrame(per_capita_access_interpolated['Country_Code'])
access_delta['Delta'] = per_capita_access_interpolated['2020']-per_capita_access_interpolated['1990']
access_delta
```

```python
delta_against_investment = energy_grid_by_country.merge(access_delta,on='Country_Code')
delta_against_investment
```

```python
plt.figure(figsize=(15,15))
plt.scatter(delta_against_investment['Amount_in_usd'],delta_against_investment['Delta'])
plt.xlabel('Amount lent in Energy and Grid in USD')
plt.ylabel('Access to electricity percentage delta, 1990 to 2020')
plt.savefig('blah.png')
plt.show()
```

```python
delta_against_investment.to_csv(os.path.join(DATA_DIR,'energy_grid_invest_vs_percent_difference.csv'))
```
