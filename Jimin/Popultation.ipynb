```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_pickle('../Extracted_Information.dat')
population = pd.read_csv('../populationdata.csv')
```

```python
def compute_WB_data(data,wbdata):
    value = []
    for i in data.index:
        country = data.Country3code[i]
        time = data.Date[i].year
        val = 0
        for j in wbdata.index:
            if wbdata.iloc[j,1] == country:
                for col in wbdata.columns:
                    if wbdata[col].iloc[3] == time:
                        val = wbdata[col].iloc[j]
                        break
        value.append(val)
    return value
```

```python
pop_value = compute_WB_data(data,population)
```

```python
a = 0
for i in range(len(pop_value)):
    if pop_value[i] == 0:
        a = a+1
        pop_value[i] = None
```

```python
a
```

```python
data["Population"] = pd.Series(pop_value)
```

```python
gdp_per_capita = []
for i in data.index:
    if data["GDP_historic"].iloc[i] != None:
        x = data["GDP_historic"].iloc[i]/data["Population"].iloc[i]
        gdp_per_capita.append(x)
    else:
        gdp_per_capita.append(None)
```

```python
gdp_per_capita
```

```python
data["GDP_per_capita"] = pd.Series(gdp_per_capita)
```
