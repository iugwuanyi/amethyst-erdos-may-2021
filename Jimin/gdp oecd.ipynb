```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_pickle('../Extracted_Information.dat')
worldmap = pd.read_csv('../world_bank_country_mappings.csv')
gdp_historical = pd.read_csv('../gdp91-21.csv')
```

```python
gdp_value = []
for i in data.index:
    country = data.Country3code[i]
    time = data.Date[i].year
    gdp = 0
    for j in gdp_historical.index:
        if gdp_historical.LOCATION[j] == country:
            for k in range(j,j+30):
                if gdp_historical.TIME[k] == time:
                    gdp = gdp_historical.Value[k]
                    break
    gdp_value.append(gdp)
               
                    
```

```python
len(gdp_value)
```

```python
gdp_value
```

```python
for i in range(len(gdp_value)):
    if gdp_value[i] == 0:
        gdp_value[i] = None
```

```python
data["GDP_historic"] = pd.Series(gdp_value)
```

```python
data.to_csv('gdp.csv')
```
